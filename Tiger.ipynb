# Image Segmentation

```python
#Import libraries
import numpy as np
import sklearn
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from matplotlib.image import imread
```

```python
#Import dataset
img = imread('tiger.png') 
plt.imshow(img)
```

```python
#Check the dimension of the image
img.shape
```

Images are in 3D format. We need to convert it to 2D for further processing.

```python
#Converting image to 2D array
new_tig = img.reshape(-1,3)
new_tig.shape
```

```python
#Fit K-Means clustering with k=16 to keep just 16 colors in the compressed image.
km = KMeans(n_clusters=16)
km.fit(new_tig)
```

```python
#Reshape the image to its original dimension
img_seg = km.cluster_centers_
img_seg = img_seg[km.labels_]
img_seg = img_seg.reshape(img.shape)
```

```python
#Display the final image
plt.imshow(img_seg)
```
